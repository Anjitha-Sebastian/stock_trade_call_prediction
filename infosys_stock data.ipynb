#### Volume Weighted Average Price

```python
#function to calculate monthwise Volume weighted average
def vvap(data):
    v=sum(data['Close Price']*data['Total Traded Quantity'])/sum(data['Total Traded Quantity'])
    return(v)
```

```python
vwap=gk.apply(vvap)
```

```python
vwap
```

```python
vwap.max()
```

```python

plt.figure(figsize=(10, 8))
vwap.plot()
plt.ylabel("VWAP",size=15)
plt.xlabel("Month,Year",size=15)
plt.title("Monthwise Volume Weighted Average Price",size=16)

plt.show()
```

```python
#average price over last N days
def avg_price_n_days(days):
    avg=sum(data['Close Price'].tail(n=days))/days
    return avg
```

```python
avg_price_n_days(7),avg_price_n_days(14),avg_price_n_days(30),avg_price_n_days(90),avg_price_n_days(180),avg_price_n_days(365)
```

#### finding profit or loss percentage change over last N days

```python

def perc_change_n_days(days):
    perc_change=data['Close Price'].tail(n=days).pct_change()*100
    l=perc_change.to_list()
    del l[0]
    avrg=sum(l)/days
    return avrg
```

```python
perc_change_n_days(7),perc_change_n_days(14),perc_change_n_days(30),perc_change_n_days(90),perc_change_n_days(180),perc_change_n_days(365)
```

### Defining trend based on daily percentage change


```python
def trend(data):
    if -0.5<data['Day_Perc_change']<0.5:
        return 'slight'
    elif 0.5<data['Day_Perc_change']<1:
        return 'slight positive'
    elif -1<data['Day_Perc_change']<-0.5:
        return 'slight negative'
    elif 1<data['Day_Perc_change']<3:
        return 'positive'
    elif -3<data['Day_Perc_change']<-1:
        return 'negative'
    elif 3<data['Day_Perc_change']<7:
        return 'Among top gainers'
    elif -7<data['Day_Perc_change']<-3:
        return 'Among top losers'
    elif data['Day_Perc_change']<-7:
        return 'bull drop'
    elif data['Day_Perc_change']>-7:
        return 'bull run'
    else:
        return'nan'
```

```python
data['Trend']=data.apply(trend, axis=1) #adding new column Trend based on conditions defied in the above function based on column Day_Perc_change
```
