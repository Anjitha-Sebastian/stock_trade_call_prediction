```python
from nsepy import get_history
from datetime import date

```

```python
bajaj = get_history(symbol='BAJAJ-AUTO',
                   start=date(2015,1,1),
                   end=date(2016,12,31))
```

```python

maruti =get_history(symbol='MARUTI',
                   start=date(2015,1,1),
                   end=date(2016,12,31))
```

```python
wipro = get_history(symbol='WIPRO',
                   start=date(2015,1,1),
                   end=date(2016,12,31))
```

```python
icic = get_history(symbol='ICICIBANK',
                   start=date(2015,1,1),
                   end=date(2016,12,31))
```

```python
reliance = get_history(symbol='RELIANCE',
                   start=date(2015,1,1),
                   end=date(2016,12,31))
```

```python
nifty = get_history(symbol="NIFTY", 
                    start=date(2015,1,1), 
                    end=date(2016,1,12),
					index=True)
```

```python
reliance.Series.unique(),icic.Series.unique(),maruti.Series.unique(),wipro.Series.unique(),bajaj.Series.unique()
```

```python
headers=['Bajaj','Maruti','Wipro','ICICI','Reliance','nifty']
closeprice=pd.concat((bajaj['Close'],maruti['Close'],wipro['Close'],icic['Close'],reliance['Close'],nifty['Close']),axis=1,keys=headers)
```

```python
closeprice
```

```python
percent_change=closeprice.pct_change()*100
```

```python
percent_change=percent_change.fillna(0)
```

```python
percent_change
```

```python
sn.pairplot(percent_change)
plt.show()
```

```python
bajaj_moving_21=bajaj['Close'].rolling(window=21).mean()
bajaj_moving_34=bajaj['Close'].rolling(window=34).mean()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

plt.plot(bajaj_moving_21,label='SMA 21')
plt.plot(bajaj_moving_34,label='SMA 34')
plt.plot(bajaj['Close'],label='Price')
plt.legend(loc=1)
ax.set_xlabel("Date",size=18)
ax.set_ylabel("Price in Rupees")
plt.tight_layout()
plt.xticks(plt.xticks()[0], rotation=45)

plt.show()
```

```python
#Task 2.10
```

```python
bajaj['MA14'] = bajaj['Close'].rolling(window=14).mean()
bajaj['14dSTD'] = bajaj['Close'].rolling(window=14).std() 

bajaj['Upper'] = bajaj['MA14'] + (bajaj['14dSTD'] * 2)
bajaj['Lower'] = bajaj['MA14'] - (bajaj['14dSTD'] * 2)
```

```python
bajaj
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

plt.plot(bajaj['MA14'],label='SMA14')
plt.plot(bajaj['Upper'],label='upper')
plt.plot(bajaj['Lower'],label='lower')
plt.plot(bajaj['Close'],label='Price',color='black')

plt.legend(loc=1)
ax.set_xlabel("Date",size=18)
ax.set_title('Bollingers Band',size=25)
plt.tight_layout()
plt.xticks(plt.xticks()[0], rotation=45)

plt.show()
```
